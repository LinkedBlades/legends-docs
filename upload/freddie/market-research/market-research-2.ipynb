```python
# social media

data = {
    "None": 36, 
    "Twitter": 43, 
    "Discord": 35.5, 
    "Youtube": 4.9, 
    "TikTok": 4.9,
    "Facebook": 3.6,
    "Twitch": 2,
    "Instagram": 1
}
X = list(data.keys())
Y = list(data.values())

plt.figure(figsize=(8,4))
fig, ax = plt.subplots()

formatter = ticker.FormatStrFormatter('%1.0f%%')
ax.xaxis.set_major_formatter(formatter)

ax.barh(X,Y)
ax.set_title("Social Media Engagement")
plt.tight_layout()
plt.savefig(os.path.join(DOCS, "11501", "social-media.svg"))
```
